## Feature Enginnering
* the features we are going to use are the recency, frequency and the monetary focus (RFM) 

```python
cleaned_df['SalesLineTotal'] = cleaned_df['Quantity']*cleaned_df['Price']
cleaned_df
```

```python
aggregated_df = cleaned_df.groupby(by= 'Customer ID', as_index=False) \
    .agg(
        MonetaryValue = ('SalesLineTotal', 'sum'),
        Frequency = ('Invoice', 'nunique'),
        LastInvoiceDate = ('InvoiceDate', 'max')
    )
aggregated_df.head(10)
```

To determine our recency, Recency refers to how many days since the clent made a purchase? its good to use today's date but it will be wierd for our data since this was data from 2011 so we are going to use the max date 

```python
max_invoice_date = aggregated_df['LastInvoiceDate'].max()
max_invoice_date
```

```python
aggregated_df['Recency']= (max_invoice_date- aggregated_df['LastInvoiceDate']).dt.days

aggregated_df.head(10)
```

with the boxplot we can see that there are outliers mostly in the monetary and frequency this outliers are important but inour clustering we will have to divide the data and remove the outliers for a separate analysis.  
the boxplot uses the quartile range in its analysis so we are going to use it in separating our outliers

```python
M_Q1= aggregated_df['MonetaryValue'].quantile(0.25)
M_Q3= aggregated_df['MonetaryValue'].quantile(0.75)
M_IQR= M_Q3 - M_Q1

#get rid of outliers
monetary_outliers_df = aggregated_df[(aggregated_df['MonetaryValue']> (M_Q3 + 1.5* M_IQR)) | (aggregated_df['MonetaryValue']< (M_Q1 - 1.5* M_IQR))].copy()

monetary_outliers_df.describe()
```

```python
F_Q1= aggregated_df['Frequency'].quantile(0.25)
F_Q3= aggregated_df['Frequency'].quantile(0.75)
F_IQR= F_Q3 - F_Q1

#get rid of outliers
frequency_outliers_df = aggregated_df[(aggregated_df['Frequency']> (F_Q3 + 1.5* F_IQR)) | (aggregated_df['Frequency']< (F_Q1 - 1.5* F_IQR))].copy()

frequency_outliers_df.describe()
```

```python
#look at the non_outliers
non_outliers_df = aggregated_df[(~aggregated_df.index.isin(monetary_outliers_df.index))& (~aggregated_df.index.isin(frequency_outliers_df.index))]

non_outliers_df.describe()
```

## KMeans clustering

```python
max_k = 12 

inertia = []
silhouette_scores = []


k_values = range(2, max_k + 1)

for k in k_values:

    kmeans = KMeans(n_clusters=k , random_state= 42, max_iter=1000)

    cluster_labels= kmeans.fit_predict(scaled_data_df)
    sil_score = silhouette_score(scaled_data_df, cluster_labels)

    inertia.append(kmeans.inertia_) 
    silhouette_scores.append(sil_score)

plt.figure(figsize=(14,6)) 


plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker = "o")
plt.title('kmean inertia for different clusters')
plt.xlabel('number of cluster(k)')
plt.ylabel('inertia')
plt.xticks(k_values)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker = "o",color= 'red')
plt.title('silhoutte scores for different values of k')
plt.xlabel('number of cluster(k)')
plt.ylabel('silhoutte scores')
plt.xticks(k_values)
plt.grid(True)

plt.tight_layout()
plt.show()
```

we look at the "elbow" and it in a point btwn the k = 4 and k = 5  
To help us know which point to choose we look at the silhoutte score.
silhoutte coefficient should vary btwn -1 to 1. coefficient close to -1 = instance is very wel inside the cluster and far from other clusters    
close to 0 = instance is close to the cluster boundary     
close to +1 = instance may be assigned to the wrong cluster   
high silhoutte score is good and a smaller one is bad(overlap is present)

based on our silhoutte score we can that choosing k =4 is much better than k = 5

```python
kmeans= KMeans(n_clusters=4, random_state=42, max_iter=1000)

cluster_labels = kmeans.fit_predict( scaled_data_df)

cluster_labels
```

```python
non_outliers_df['cluster']= cluster_labels

non_outliers_df.head(10)
```

```python
# lets look at the outlier data 
overlap_indices = monetary_outliers_df.index.intersection(frequency_outliers_df.index)

monetary_only_outliers = monetary_outliers_df.drop(overlap_indices)
frequency_only_outliers = frequency_outliers_df.drop(overlap_indices)

monetary_and_frequency_outliers = monetary_outliers_df.loc[overlap_indices]

#lets manually assign the clusters in the data
monetary_only_outliers['cluster']=  -1
frequency_only_outliers['cluster']= -2
monetary_and_frequency_outliers['cluster']= -3

outlier_cluster_df = pd.concat([monetary_only_outliers, frequency_only_outliers, monetary_and_frequency_outliers])
outlier_cluster_df
```

```python
cluster_colors = {-1: '#ff00ff',  # pink
                  -2: '#40ff00',  # Green
                  -3: '#000000'}  # black

fig = plt.figure(figsize=(12,18))

plt.subplot(3, 1, 1)
sns.violinplot(x= outlier_cluster_df['cluster'], y= outlier_cluster_df['MonetaryValue'], palette=cluster_colors, hue=outlier_cluster_df['cluster'])
sns.violinplot(y= outlier_cluster_df['MonetaryValue'], color='gray', linewidth=1.0)
plt.title('monetary value by cluster')
plt.ylabel('monetary value')

plt.subplot(3, 1, 2)
sns.violinplot(x=outlier_cluster_df['cluster'], y=outlier_cluster_df['Frequency'], palette=cluster_colors, hue= outlier_cluster_df['cluster'])
sns.violinplot(y=outlier_cluster_df['Frequency'], color='gray', linewidth=1.0)
plt.title('Frequency  by cluster')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
sns.violinplot(x=outlier_cluster_df['cluster'], y=outlier_cluster_df['Recency'], palette=cluster_colors, hue= outlier_cluster_df['cluster'])
sns.violinplot(y= outlier_cluster_df['Recency'], color='gray', linewidth=1.0)
plt.title('Recency  by cluster')
plt.ylabel('Recency')

plt.tight_layout()
plt.show()
```

* Cluster -1 (Monetary Outliers) PAMPER: Characteristics: High spenders but not necessarily frequent buyers. Their purchases are large but infrequent. Potential Strategy: Focus on maintaining their loyalty with personalized offers or luxury services that cater to their high spending capacity.
* Cluster -2 (Frequency Outliers) UPSELL: Characteristics: Frequent buyers who spend less per purchase. These customers are consistently engaged but might benefit from upselling opportunities.
```python
cluster_labels = {
    0: "RETAIN",
    1: "RE-ENGAGE",
    2: "NURTURE",
    3: "REWARD",
    -1: "PAMPER",
    -2: "UPSELL",
    -3: "DELIGHT"
}
```
